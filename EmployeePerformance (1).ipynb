```python
#Read the analytics csv file and store our dataset into a dataframe called "df"
df = pd.read_csv('Employee_Performance.csv')
```

```python
# Create a new column 'AgeGroup' in the DataFrame
age_bins = [0, 25, 35, 45, 55, 100]
age_labels = ['<25', '25-35', '35-45', '45-55', '55+']
df['AgeGroup'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)
```

# visualization of Data

```python
#Gender Distribution Among Employees
colors = ['#1f77b4', '#f0b27a']
ax = sns.countplot(x='Gender',data =df,palette='viridis')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
#Distribution of Age
mean_age = df['Age'].mean()
mean_age_formatted = f'{mean_age:.2f}'
plt.figure(figsize=(8, 5))
sns.kdeplot(x=df['Age'], color='tomato',shade=True,label='Age')
plt.axvline(x=mean_age,color ='Purple',linestyle='--',label=f'Mean Age:{mean_age_formatted}')
plt.legend()
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

##### The peak (highest density) occurs around the age of 30-35, suggesting that most individuals in this group are within this age range.

```python
#Attrition Male And Female
fig,axes = plt.subplots(1,2,sharex=True,figsize=(15,5))
fig.suptitle('Attrition Age Distribution by Gender')
sns.kdeplot(ax=axes[0],x=df[(df['Gender']=='Male')&(df['Attrition']=='Yes')]['Age'],color='purple',shade=True,label='Yes')
sns.kdeplot(ax=axes[0],x=df[(df['Gender']=='Male')&(df['Attrition']=='No')]['Age'],color='green',shade=True,label='No')
axes[0].set_title('Male')
axes[0].legend(title='Attrition')
sns.kdeplot(ax=axes[1],x=df[(df['Gender']=='Female')&(df['Attrition']=='Yes')]['Age'],color='purple', shade=True,label='Yes')
sns.kdeplot(ax=axes[1],x=df[(df['Gender']=='Female')&(df['Attrition']=='No')]['Age'],color='green',shade=True,label='No')
axes[1].set_title('Female')
axes[1].legend(title='Attrition')
plt.show()
```

##### Both male and female employees who are younger, particularly in their early 30s, seem more prone to attrition. In contrast, employees who are slightly older, in their mid to late 30s, tend to stay with the organization.

```python
# Agewise PerformanceRating of People in an Organization
age_att = df.groupby(['Age','PerformanceRating']).size().reset_index(name='Counts')
plt.figure(figsize=(14, 6))
sns.lineplot(data=age_att, x='Age',y='Counts',hue='PerformanceRating',marker='o')
plt.title('Agewise PerformanceRating of People in an Organization')
plt.xlabel('Age')
plt.ylabel('Counts')
plt.legend(title='PerformanceRating')
plt.show()
```

##### Rating 3 is the most common performance rating, especially for people in their 30s and 40s. There’s a notable peak around ages 30-40, after which the count gradually decreases.
##### Rating 2 shows some fluctuation, but overall, the counts are relatively low compared to Rating 3. The counts rise slightly in the 20s and 30s age range, with some minor peaks, and then decline.
##### Rating 4, the highest performance rating, has consistently low counts across all age groups, indicating that this rating is rare regardless of age.

```python
# Employee performance ratings Distribution
plt.figure(figsize=(8,4))
ax = sns.countplot(x='PerformanceRating',data=df,palette='viridis')

for bars in ax.containers:
  ax.bar_label(bars)
plt.title('Distribution of Performance Ratings Among Employees')
plt.xlabel('Performance Rating')
plt.ylabel('Count')
plt.show()
```

##### Rating 3 is the most common performance rating, with 874 employees falling into this category. This suggests that the majority of the workforce is performing at an average or satisfactory level.
##### Rating 4 is the least common, with only 132 employees achieving this rating, indicating that high performance is relatively rare in this organisation.
##### Rating 2, while less common than Rating 3, still has a significant number of employees (194) performing below the average.

```python
#Performnace Affecting Age Group
plt.figure(figsize=(8,4))
ax =sns.countplot(data =low_performance_df,x ='AgeGroup',hue ='Gender',palette='viridis')

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
#EductaionField of Employees
plt.figure(figsize=(10,5))
ax =sns.countplot(x='EducationBackground',data=low_performance_df,palette='viridis')
plt.title('Distrubtion Of EducationField Freqency BusinessTravel',fontsize=10)

for bars in ax.containers:
  ax.bar_label(bars)
```

##### Life Sciences and Medical backgrounds dominate the group of frequent business travelers, which could indicate that these fields are more likely to require travel, perhaps due to industry demands, client interactions, or sales roles.

```python
#Experinece of YearsInCurrentRole
plt.figure(figsize=(8,4))
ax =sns.countplot(data =low_performance_df, x ='ExperienceYearsInCurrentRole',palette='viridis')

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
#Employee Performance rating Analysis 
att1= df.groupby('PerformanceRating',as_index=False)['Age'].count()
att1['Count']= att1['Age']
att1.drop('Age', axis=1, inplace=True)
att2 = df.groupby(['Gender','PerformanceRating'],as_index=False)['Age'].count()
att2['Count']= att2['Age']
att2.drop('Age',axis=1,inplace=True)
colors = ['#636EFA','#EF553B','#00CC96','#AB63FA','#FFA15A']
fig,axes= plt.subplots(1, 3, figsize=(18, 6))

# Plot for overall Employee PerformanceRating
axes[0].pie(att1['Count'],labels=att1['PerformanceRating'],autopct='%1.1f%%',colors=colors,startangle=140)
axes[0].set_title('Employee PerformanceRating')

# Plot for Female PerformanceRating
female_data=att2[att2['Gender'] =='Female']
axes[1].pie(female_data['Count'],labels=female_data['PerformanceRating'],autopct='%1.1f%%',colors=colors,startangle=140)
axes[1].set_title('Female PerformanceRating')

# Plot for Male PerformanceRating
male_data=att2[att2['Gender'] == 'Male']
axes[2].pie(male_data['Count'],labels=male_data['PerformanceRating'],autopct='%1.1f%%',colors=colors,startangle=140)
axes[2].set_title('Male PerformanceRating')
fig.suptitle('Performance Rating Analysis',fontsize=16)
plt.show()
```

##### The performance rating distribution is quite consistent across genders, with the majority of employees, regardless of gender, receiving a rating of 3
##### The organization appears to have a balanced performance rating system with similar evaluation outcomes for both male and female employees
